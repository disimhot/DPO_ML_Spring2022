```python
correlated_features = set()
for i in range(len(mat.columns)):
    for j in range(i):
        if abs(mat.iloc[i, j]) > 0.8:
            colname = mat.columns[i]
            correlated_features.add(colname)

correlated_features
```

```python
X.drop(labels=correlated_features, axis=1, inplace=True)
test.drop(labels=correlated_features, axis=1, inplace=True)
```

```python
st = make_column_transformer(
            (
                Pipeline(steps = [
                        ("s1", SimpleImputer(missing_values=np.nan, strategy="median")),
                        ("s2", StandardScaler()),
                       
                ]),                                     
                make_column_selector(dtype_include=np.number)
            ),
            (
                OneHotEncoder(handle_unknown='ignore'), make_column_selector(dtype_include=object)
            )
)

# parameters combinations
params = {"classifier__n_estimators": [10, 50, 100],
          "classifier__max_depth": [10, 30, 100],
         }


pipeline = Pipeline([
    ('scale-transform', st),
    ('feature_selection', SelectFromModel(LinearSVC())),
    ('classifier', RandomForestRegressor())
])

grid = GridSearchCV(estimator=pipeline, param_grid=params)

grid.fit(X, y)
print(grid.best_params_)
```

```python
# обучим модель с этими параметрами
model_rand_forest = Pipeline([
    ('scale-transform', st),
    ('feature_selection', SelectFromModel(LinearSVC())),
    ('classifier', RandomForestRegressor(n_estimators=10, max_depth=10))
])

model_rand_forest.fit(X, y)

train_pred = model_rand_forest.predict(X)

print("Train MSE: {}".format(np.sqrt(mean_squared_error(y, train_pred))))
print(f'''train R2={r2_score(y, train_pred)}''')
```

```python
# используем cat boosting
model_cat_boost = Pipeline([
    ('scale-transform', st),
    ('feature_selection', SelectFromModel(LinearSVC())),
    ('classifier', CatBoostRegressor())
])

model_cat_boost.fit(X, y)

train_pred = model_cat_boost.predict(X)

print("Train MSE: {}".format(np.sqrt(mean_squared_error(y, train_pred))))
print(f'''train R2={r2_score(y, train_pred)}''')
```

```python
test_id=test['Id']
X_predict = test.drop('Id', axis=1)
y_pred = model_cat_boost.predict(X_predict)


submission = pd.DataFrame({
    'Id': test_id,
    'SalePrice': y_pred
})

submission
```
