```python
# статистические разбросы данных
def summary_statistics(df):
    len_df = len(df)
    seria = pd.concat([df.dtypes, df.isnull().sum(),(df==0).sum()], axis=1, keys=['dtypes', 'Число_пропущ','Число_нулей'])
    seria['Доля_пропущ'] = round(100 * seria['Число_пропущ'] / len_df, 2)
    seria['Доля_нулей'] = round(100 * seria['Число_нулей'] / len_df, 2)
    desc = pd.concat([seria, df.describe(include='all').T], axis=1)
    return desc.fillna('').sort_values(by=['Число_пропущ','Число_нулей','dtypes','unique','mean'], ascending = False)
```

```python
y = train[['SalePrice']]
Xwork = train.drop(['SalePrice', 'Id'], axis=1)
Xwork
```

```python
train_stat = summary_statistics(Xwork)
train_stat
```

```python
# в качестве идеи

# разобъем признаки на
# 1) содержащие много пропусков
nan_col = train_stat[train_stat['Доля_пропущ'] >= 20].index.values

# 'Периметр_участка' - попробуем востановить моделью
spec_num_col = train_stat[(train_stat['Доля_пропущ'] >= 6) &
                          (train_stat['Доля_пропущ'] < 20)].index.values

# 2) содержащие много нулей - "разряженые"
zero_col = train_stat[train_stat['Доля_нулей'] >= 30].index.values

# 3) "хорошие"
work_col = train_stat[(train_stat['Доля_пропущ'] < 6) &
                      (train_stat['Доля_нулей'] < 30)].index.values

cat_col = train_stat.loc[work_col,][train_stat['dtypes'] == 'object'].index.values
num_col = train_stat.loc[work_col,][train_stat['dtypes'] != 'object'].index.values

if 0==len(Xwork.columns) - len(nan_col) - len(spec_num_col) - len(zero_col) - len(cat_col) - len(num_col):
    print('ok')
    print(len(nan_col))
else:
    print('!ОШИБКА! - несогласованное число признаков')
```

```python
from category_encoders.one_hot import OneHotEncoder
from category_encoders.target_encoder import TargetEncoder
from category_encoders.leave_one_out import LeaveOneOutEncoder
from category_encoders.quantile_encoder import QuantileEncoder

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python

cat_processor = Pipeline([ #OneHotEncoder(drop='if_binary')
    ('Заполнение_пропусков', SimpleImputer(strategy="most_frequent", add_indicator=True)),
    ('Кодер_категорий', TargetEncoder()), #
    ('Стандартизация', StandardScaler())
])

num_processor_spec = Pipeline([
    ('Восстановление_пропусков' , IterativeImputer(random_state=42)), # востановление пропусков
    # ('Заполнение_пропусков', SimpleImputer(strategy="mean", add_indicator=True)),
    ('Стандартизация', StandardScaler())
])

num_processor_disp = Pipeline([
    ('Очищение_дисперсии', VarianceThreshold(threshold=10.0)),
    ('Заполнение_пропусков', SimpleImputer(strategy="mean", add_indicator=True)),
    ('Мин_Макс_стандартизация', MinMaxScaler())
])

num_processor = Pipeline([
    ('Заполнение_пропусков', SimpleImputer(strategy="mean", add_indicator=True)),
    ('Стандартизация_значений', StandardScaler())
])
```

```python
# обработаем группы признаков
column_processor = ColumnTransformer([
    ('Кодирование_категорий', cat_processor, cat_col),
    ('Восстановление_числ_признака', num_processor_spec, spec_num_col),
    ('Обработка_разряженых_признаков', num_processor_disp, zero_col),
    ('Обработка_общая_признаков', num_processor, num_col)
], remainder='drop')
```

```python
model1 = Pipeline([
    ('Подготовка_признаков', column_processor),
    # ('Рекурсивное_исключение',  RFE(LogisticRegression(),  step=1)),
    ('Лес_деревьев',RandomForestRegressor(oob_score = True)) # гипер параметров нет
])

model2 = Pipeline([
    ('Подготовка_признаков', column_processor),
    ('Градиентный_бустинг', GradientBoostingRegressor()) # гипер параметров нет
])
```

```python
tt = TransformedTargetRegressor(regressor=model2, func=np.log, inverse_func=np.exp)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train.drop('SalePrice', axis=1), y, test_size=0.3, shuffle=True, random_state=42)
```

```python
tt.fit(X_train, y_train)
tt.regressor_[-1].train_score_.mean()
```

```python
cv_res = cross_validate(tt,
                            X_train,
                            y_train,
                            cv=5,
                            scoring=['r2', 'neg_mean_squared_log_error']
                           )
print(f"Оценка Кросс-валидацией Rкв = {np.median(cv_res['test_r2']).round(2)}")
print(f"Оценка Кросс-валидацией RMSE = {-np.median(cv_res['test_neg_mean_squared_log_error']).round(4)}")
```

```python
print("r2 test: ", r2_score(y_test, tt.predict(X_test)).round(2))
print('rmse test: ', mean_squared_log_error(y_test, tt.predict(X_test)).round(4))
```

```python
# поиск гиперпараметров по сетке 
# считается порядка 10 минут

param_grid = dict(
    Градиентный_бустинг__max_depth =[3, 4, 5, 6],
    Градиентный_бустинг__n_estimators =[100, 300, 500],
    Градиентный_бустинг__learning_rate =[0.1, 0.5, 0.01],
)

grid_search = GridSearchCV(model2, param_grid=param_grid, verbose=5)
grid_search.fit(X_train, np.log(y_train.to_numpy()))
print(grid_search.best_params_)
```

```python
dict_best_params = grid_search.best_params_
dict_best_params
```

```python
# УСТАНОВИМ ЛУЧШИЕ ПАРАМЕТРЫ
tt.regressor_.set_params(**dict_best_params)

# ОБУЧАЕМСЯ УЖЕ НА ВСЕЙ ВЫБОРКЕ:
tt.fit(train.drop('SalePrice', axis=1), y)

print("r2 test: ", r2_score(y_test, tt.predict(X_test)).round(2))
print('rmse test: ', mean_squared_log_error(y_test, tt.predict(X_test)).round(4))

y_pred = tt.predict(test)

submission = pd.DataFrame({
    'Id': test.Id,
    'SalePrice': y_pred[:,0]
})

submission.head()
```
