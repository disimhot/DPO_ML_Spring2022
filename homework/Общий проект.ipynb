```python
# статистические разбросы данных
def summary_statistics(df):
    len_df = len(df)
    seria = pd.concat([df.dtypes, df.isnull().sum(),(df==0).sum()], axis=1, keys=['dtypes', 'Число_пропущ','Число_нулей'])
    seria['Доля_пропущ'] = round(100 * seria['Число_пропущ'] / len_df, 2)
    seria['Доля_нулей'] = round(100 * seria['Число_нулей'] / len_df, 2)
    desc = pd.concat([seria, df.describe(include='all').T], axis=1)
    return desc.fillna('').sort_values(by=['Число_пропущ','Число_нулей','dtypes','unique','mean'], ascending = False)
```

```python

cat_processor = Pipeline([ #OneHotEncoder(drop='if_binary')
    ('Заполнение_пропусков', SimpleImputer(strategy="most_frequent", add_indicator=True)),
    ('Кодер_категорий', TargetEncoder()), #
    ('Стандартизация', StandardScaler())
])

num_processor_spec = Pipeline([
    ('Восстановление_пропусков' , IterativeImputer(random_state=42)), # востановление пропусков
    # ('Заполнение_пропусков', SimpleImputer(strategy="mean", add_indicator=True)),
    ('Стандартизация', StandardScaler())
])

num_processor_disp = Pipeline([
    # ('Очищение_дисперсии', VarianceThreshold(threshold=10.0)),
    ('Заполнение_пропусков', SimpleImputer(strategy="mean", add_indicator=True)),
    ('стандартизация', MinMaxScaler())
])

num_processor = Pipeline([
    ('Заполнение_пропусков', SimpleImputer(strategy="mean", add_indicator=True)),
    ('Стандартизация_значений', StandardScaler())
])
```
